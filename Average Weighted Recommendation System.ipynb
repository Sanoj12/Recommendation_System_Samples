```python
import pandas as pd
import numpy as np
```

```python
#https://www.kaggle.com/tmdb/tmdb-movie-metadata
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.head()
```

```python
movies_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies_df.head()
```

```python
##shape of dataset

print("credits:",credits.shape)
print("movie dataframe:",movies_df.shape)
```

```python
##combine both dataset
credits_column_renamed = credits.rename(index=str,columns={"movie_id":"id"})

                                        
credits_column_renamed
```

```python
movies_df_merge =  movies_df.merge(credits_column_renamed, on='id') 
```

```python
movies_df_merge.head()
```

```python
movies_df_cleaned = movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
```

```python
movies_df_cleaned.head()
```

```python
movies_df_cleaned.info()
```

```python
####calculate weighted 
v=movies_df_cleaned['vote_count']
R = movies_df_cleaned['vote_average']
C = movies_df_cleaned['vote_average'].mean()
m = movies_df_cleaned['vote_count'].quantile(0.70)

```

```python
movies_df_cleaned['weighted_average'] = ((R*v) + (C*m) )/ (v+m)
```

```python
movies_df_cleaned.head()
```

```python
##weighted average order

movie_sorted_ranking = movies_df_cleaned.sort_values('weighted_average',ascending=False)
movie_sorted_ranking [['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average = movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))

axis1 = sns.barplot(x=weight_average['weighted_average'].head(10) , y= weight_average['original_title'].head(10),data=weight_average)

plt.xlim(4,10)
plt.title("Best Movie by average Votes",weight='bold')
plt.xlabel('weighted average score', weight='bold')
plt.ylabel('Movie title', weight='bold')
plt.savefig('best_movies.png')
plt.show()
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

#####RECOMMENTATION BASED ON SCALED WEIGHTED AVERAGE AND POPULARITY SCORE(BOTH ARR SAME PRIORITY0)


```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()

movies_df_scaled = scaling.fit_transform(movies_df_cleaned[['weighted_average','popularity']])




                                       
```

```python
movies_df_normalized  =  pd.DataFrame(movies_df_scaled,columns=['weighted_average','popularity'])
```

```python
movies_df_normalized.head()
```

```python
movies_df_cleaned[['normalized_weight_average','normalized_popularity']]= movies_df_normalized
```

```python
movies_df_cleaned.head()
```

```python
movies_df_cleaned['score'] = movies_df_cleaned['normalized_weight_average'] * 0.5   +   movies_df_cleaned['normalized_popularity'] * 0.5
```

```python
movies_df_scored = movies_df_cleaned.sort_values(['score'] ,ascending=False)
movies_df_scored[['original_title','normalized_weight_average','normalized_popularity','score']]
```
